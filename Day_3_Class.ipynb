```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
sns.set(style = 'darkgrid')
tips = sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
tips.describe()
```

```python
tips['size']
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, style = 'size')
```

```python
from numpy.random import randn
```

```python
df = pd.DataFrame(dict(time = np.arange(500), value = randn(500).cumsum()))
```

```python
df.head()
```

```python
sns.relplot(x = 'time', y = 'value', data = df, kind = 'line', sort = True)
```
